```python
pc1=read.groupby(['STATES','SIZE']).agg({'PRICE_MEDIAN':'median','COUNTS':'sum'})
pc1
```

```python
pc1=pc1.reset_index()
```

```python
pc1.drop(columns='PRICE_MEDIAN',inplace=True)
pc1=pc1.groupby(['STATES'])[['COUNTS']].sum()
pc1.head(5)
```

```python
index=pc1.index[:5]
index
```

```python
con=read['STATES'].isin(index)
read=read.loc[con]
read
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib import rc
rc('font',family='NanumGothic')
sns.violinplot(data=read,x='SIZE',y='PRICE_MEDIAN',hue='STATES')
plt.title('자치구와 규모에따른 가격밀도표')
plt.xlabel('size')
plt.legend(loc='upper right')
plt.xticks()
plt.yticks()
plt.grid(True)
plt.show()
```

```python
sns.histplot(
    data=read,
    x="PRICE_MEDIAN",
    hue="STATES",
    bins=30,
    kde=True
)
```

```python
sns.histplot(data=read, x="PRICE_MEDIAN",hue='STATES', bins=20,kde=False)
```

```python
sns.ecdfplot(
    data=read,
    x="PRICE_MEDIAN",
    hue="STATES"
)
```

```python
import numpy as np

sns.barplot(
    data=read,
    x="SIZE",
    y="PRICE_MEDIAN",
    hue="STATES",
    estimator=np.median,
    errorbar=None
)
```

```python
sns.stripplot(
    data=read,
    x="SIZE",
    y="PRICE_MEDIAN",
    hue="STATES",
    dodge=True,
    alpha=0.4
)
```
