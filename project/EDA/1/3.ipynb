```python
import pandas as pd  


read=pd.read_csv('/home/hyuksu/projects/ml/project/데이터/FIN_DATA_v5.csv',encoding='utf-8')
read
```

```python
read.sort_values('DATE_YM',ascending=False).reset_index(inplace=True)
read
```

```python
con1=read['STATES']=='광진구'
read=read[con1]
read['DATE_YM']=pd.to_datetime(read['DATE_YM'].astype(str),format='%Y%m')
read
```

```python
read_copy=read.copy()
size=read['SIZE'].unique()
size.sort()
```

```python
size=list(size)
size.remove('대형')
size.append('대형')
size
```

```python
import matplotlib.pyplot as plt 
import matplotlib.dates as mdates
from matplotlib import rc
plt.figure(figsize=(12,5))
rc('font', family='NanumGothic')
for i in size:
    con = read_copy['SIZE'].eq(i)                 # ✅ read_copy 기준
    df_i = read_copy.loc[con].sort_values('DATE_YM')
    plt.plot(df_i['DATE_YM'], df_i['PRICE_MEDIAN'], marker='', label=f'광진구_{i}')

ax = plt.gca()
ax.xaxis.set_major_locator(mdates.YearLocator())
ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y'))
ax.xaxis.set_minor_locator(mdates.MonthLocator())

plt.ylabel('금액')
plt.xlabel('년도')
plt.title('광진구 SIZE별 중위가격 추이')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
read=read[read['STATES']=='강남구']
read['DATE_YM']=pd.to_datetime(read['DATE_YM'],format='%Y%m')
read
```

```python
import matplotlib.pyplot as plt 
import matplotlib.dates as mdates 
from matplotlib import rc 
rc('font',family='NanumGothic') 
read_copy=read.copy() 
fig,ax1=plt.subplots(figsize=(12,7)) 
ax2=ax1.twinx() 
for i in size: 
    temp=read_copy[read_copy['SIZE']==i] 
    temp.sort_values('DATE_YM',inplace=True) 
    ax1.plot(temp['DATE_YM'],temp['PRICE_MEDIAN'],label=f'강남구_{i}가격') 
temp1=read_copy.groupby('DATE_YM')['COUNTS'].sum().sort_index() 
ax2.plot(temp1.index,temp1.values,label='거래량',color='red') 
ax=plt.gca() 
ax.xaxis.set_major_locator(mdates.YearLocator()) 
ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y%m')) 
ax.xaxis.set_minor_locator(mdates.MonthLocator()) 
# 7) 라벨/타이틀
ax1.set_xlabel('기간')
ax1.set_ylabel('가격(PRICE_MEDIAN)')
ax2.set_ylabel('거래량(COUNTS)')
ax1.set_title('강남구 SIZE별 가격(좌) + 거래량 합계(우)')
ax1.grid(True)

# 8) 범례 합치기
h1, l1 = ax1.get_legend_handles_labels()
h2, l2 = ax2.get_legend_handles_labels()
ax1.legend(h1 + h2, l1 + l2, loc='upper left')

#fig.tight_layout()
#fig.autofmt_xdate()
plt.show()
```
