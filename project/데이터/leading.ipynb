```python
import pandas as pd   
import numpy as np 
```

```python
crime=pd.read_csv("/home/hyuksu/projects/ml/project/데이터/범죄율/5대_범죄_발생현황(2024년).csv")
crime
```

```python
crime=crime.iloc[3:].reset_index(drop=True)
crime
```

```python
df=crime.rename(columns={
    "자치구별(2)": "구",
    "2024": "살인",
    "2024.1": "강도",
    "2024.2": "강간_강제추행",
    "2024.3": "절도",
    "2024.4": "폭력"
})
df
```

```python
crime_df = df[["구", "살인", "강도", "강간_강제추행", "절도", "폭력"]]
crime_df = crime_df.replace("-", 0)
crime_df[["살인", "강도", "강간_강제추행", "절도", "폭력"]] = \
    crime_df[["살인", "강도", "강간_강제추행", "절도", "폭력"]].astype(int)
```

```python
crime_df.set_index("구",inplace=True)
crime_df
```

```python
crime_df["Crime_Score"] = (
    0.35 * crime_df["살인"]
  + 0.20 * crime_df["강도"]
  + 0.25 * crime_df["강간_강제추행"]
  + 0.10 * crime_df["절도"]
  + 0.10 * crime_df["폭력"]
)
crime_df
```
