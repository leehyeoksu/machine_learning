```python
import pandas as  pd

df=pd.read_csv('/home/hyuksu/projects/ml/project/데이터/FIN_DATA_v4.csv',encoding='utf-8')
```

```python
df.sort_values('DATE_YM', inplace=True)
df.reset_index(drop=True, inplace=True)
```

```python
df
```

```python
df.drop(['DATE_YM'], axis=1, inplace=True)
```

```python
df
```

```python
from sklearn.ensemble import AdaBoostRegressor
from sklearn.preprocessing import OneHotEncoder
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.metrics import mean_absolute_error, mean_squared_error
import numpy as np

df = df.copy()


y = df["VARIABLED"]
X = df.drop(columns=["VARIABLED"])

one_hot = X.select_dtypes(include=["object"]).columns
numeric = X.select_dtypes(exclude=["object"]).columns


preprocessor = ColumnTransformer(
    transformers=[
        ("num", "passthrough", numeric),
        ("cat", OneHotEncoder(handle_unknown="ignore"), one_hot),
    ]
)

ada_model = Pipeline(steps=[
    ("preprocessor", preprocessor),
    ("regressor", AdaBoostRegressor(n_estimators=100, random_state=42))
])

split = int(len(X) * 0.8)
train_x, test_x = X.iloc[:split], X.iloc[split:]
train_y, test_y = y.iloc[:split], y.iloc[split:]

ada_model.fit(train_x, train_y)
#train값 평가
train_pred = ada_model.predict(train_x)
train_mae = mean_absolute_error(train_y, train_pred)
train_rmse = np.sqrt(mean_squared_error(train_y, train_pred))
train_r2 = ada_model.score(train_x, train_y)
print(f"[AdaBoostRegressor] Train MAE={train_mae:.6f} RMSE={train_rmse:.6f} R2={train_r2:.6f}")

#test값 퍙가
pred = ada_model.predict(test_x)

mae = mean_absolute_error(test_y, pred)
rmse = np.sqrt(mean_squared_error(test_y, pred))
r2 = ada_model.score(test_x, test_y)

print(f"[AdaBoostRegressor] Test MAE={mae:.6f} Test_RMSE={rmse:.6f} Test_R2={r2:.6f}")
```
