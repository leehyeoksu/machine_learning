```python
import pandas as pd

reading=pd.read_csv('/home/hyuksu/projects/ml/project/데이터/FIN_DATA_v5.csv',encoding='utf-8')
read=reading.copy()
reading
```

```python
read.sort_values('DATE_YM', inplace=True)
read.drop(columns=['DATE_YM'],inplace=True)
read
```

```python
#데이터 시계열 정리 완료
read.reset_index(drop=True,inplace=True)
read
```

```python
from sklearn.svm import SVR #svr모델 부름
from sklearn.metrics import mean_squared_error # mse 측정  ytrue, ypred넣어야됨
import numpy as np # 그냥 vscode 자동완성이해줌
from sklearn.preprocessing import OneHotEncoder #원핫인코더 부름  구 하고 크기 object ->  숫자형으로 바꿔줌
from sklearn.compose import ColumnTransformer # pipline에서 규칙성 부가할거임
from sklearn.pipeline import Pipeline #파이프라인 부름
df=read.copy() # 오염 방지 복사
x=df.drop(columns=['VARIABLED']) # x값 분리
y=df['VARIABLED'] # y값만
str_type=x.select_dtypes(include=['object']).columns.to_list() # x값에 object인 타입 뽑음 onehot 전 onethor lsit
num_type=x.select_dtypes(exclude=['object']).columns.to_list() # x값에 숫자타입
length=int(len(df)*0.75) # 퍼센테이지로 크기 자름 그냥 넣으먄 인덱싱 안됨
train_x,test_x=x.iloc[:length],x.iloc[length:]
train_y,test_y=y.iloc[:length],y.iloc[length:]
preprocess=ColumnTransformer(
    transformers=[
        ('onehot', OneHotEncoder(handle_unknown='ignore'), str_type),#code컬럼에 원핫인코더 적용
        ('num', 'passthrough',num_type) #숫자형 컬럼은 그대로 둠
    ])

svr_model=Pipeline([('prep',preprocess),('svr',SVR(
    kernel='rbf',
    C=10.0,
    epsilon=0.1,
    gamma='scale'
    ))])
```

```python
# 학습
svr_model.fit(train_x,train_y)
```

```python
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
import numpy as np
train_pred=svr_model.predict(train_x)
test_pred=svr_model.predict(test_x)

#train에서 scrore보기
print('-----train')
print(f'r2_score:{r2_score(train_y,train_pred)}')
print(f'mse:{mean_squared_error(train_y,train_pred)}')
print(f'rmse:{np.sqrt(mean_squared_error(train_y,train_pred))}')
print(f'mae:{mean_absolute_error(train_y,train_pred)}')

# test에서 score 보기
print('-----test')
print(f'r2_score:{r2_score(test_y,test_pred)}')
print(f'mse:{mean_squared_error(test_y,test_pred)}')
print(f'rmse:{np.sqrt(mean_squared_error(test_y,test_pred))}')
print(f'mae:{mean_absolute_error(test_y,test_pred)}')

```
