```python
import pandas as pd

read= pd.read_csv('/home/hyuksu/projects/ml/project/데이터/FIN_DATA_v5.csv',encoding='utf-8')
read
```

```python
read=read.sort_values('DATE_YM')
read.drop(columns='DATE_YM',inplace=True)
```

```python
read.reset_index(drop=True,inplace=True)
read
```

```python
read_copy=read.copy()
y=read_copy['VARIABLED']
x=read_copy.drop(columns='VARIABLED')
```

```python
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score

str_type=x.select_dtypes(include='object').columns.to_list()
num_type=x.select_dtypes(exclude='object').columns.to_list()

pre=ColumnTransformer(transformers=[('onehot',OneHotEncoder(handle_unknown='ignore'),str_type),('num','passthrough',num_type)])

gbr=Pipeline(steps=[('prep',pre),('gbr',GradientBoostingRegressor(n_estimators=100,learning_rate=0.1,max_depth=3))])
num=int(len(read))*0.75
train_x,test_x=x.loc[:num],x.loc[num:]
train_y,test_y=y.loc[:num],y.loc[num:]

gbr.fit(train_x,train_y)
print(f'학습데이터{gbr.score(train_x,train_y)} 테스트데이터{gbr.score(test_x,test_y)}')
```

```python
train_pre=gbr.predict(train_x)
test_pre=gbr.predict(test_x)

a=mean_squared_error(train_y,train_pre)
a1=mean_absolute_error(train_y,train_pre)
b=mean_squared_error(test_y,test_pre)
b1=mean_absolute_error(test_y,test_pre)
print('train-----------')
print(f'mse:{a},mae{a1}')
print('test-----------')
print(f'mse:{b},mae{b1}')
```
