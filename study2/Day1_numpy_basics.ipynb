```python
# ë²¡í„° ë§ì…ˆ
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])

# ì›ì†Œë³„ ë§ì…ˆ
c = a + b
print("ë²¡í„° ë§ì…ˆ:")
print(f"a = {a}")
print(f"b = {b}")
print(f"a + b = {c}")
print()

# ê³„ì‚° ê³¼ì •:
# [1+4, 2+5, 3+6] = [5, 7, 9]
```

### 2.2 í–‰ë ¬ ì—°ì‚°

```python
# í–‰ë ¬ ìƒì„±
A = np.array([[1, 2],
              [3, 4]])
B = np.array([[5, 6],
              [7, 8]])

print("í–‰ë ¬ A:")
print(A)
print()
print("í–‰ë ¬ B:")
print(B)
print()
```

```python
# í–‰ë ¬ ë§ì…ˆ (ì›ì†Œë³„)
C = A + B
print("í–‰ë ¬ ë§ì…ˆ (A + B):")
print(C)
print()

# ê³„ì‚° ê³¼ì •:
# [[1+5, 2+6],  = [[6, 8],
#  [3+7, 4+8]]     [10, 12]]
```

```python
# ì›ì†Œë³„ ê³±ì…ˆ (element-wise multiplication, Hadamard product)
D = A * B
print("ì›ì†Œë³„ ê³±ì…ˆ (A * B):")
print(D)
print()

# ê³„ì‚° ê³¼ì •:
# [[1*5, 2*6],  = [[5, 12],
#  [3*7, 4*8]]     [21, 32]]
```

```python
# í–‰ë ¬ ê³±ì…ˆ (matrix multiplication)
E = A @ B  # ë˜ëŠ” np.matmul(A, B)
print("í–‰ë ¬ ê³±ì…ˆ (A @ B):")
print(E)
print()

# ê³„ì‚° ê³¼ì • (ìì„¸íˆ):
# E[0,0] = A[0,:] Â· B[:,0] = 1*5 + 2*7 = 5 + 14 = 19
# E[0,1] = A[0,:] Â· B[:,1] = 1*6 + 2*8 = 6 + 16 = 22
# E[1,0] = A[1,:] Â· B[:,0] = 3*5 + 4*7 = 15 + 28 = 43
# E[1,1] = A[1,:] Â· B[:,1] = 3*6 + 4*8 = 18 + 32 = 50
#
# ê²°ê³¼: [[19, 22],
#       [43, 50]]

print("\nìƒì„¸ ê³„ì‚° ê³¼ì •:")
print(f"E[0,0] = {A[0,0]}*{B[0,0]} + {A[0,1]}*{B[1,0]} = {A[0,0]*B[0,0]} + {A[0,1]*B[1,0]} = {E[0,0]}")
print(f"E[0,1] = {A[0,0]}*{B[0,1]} + {A[0,1]}*{B[1,1]} = {A[0,0]*B[0,1]} + {A[0,1]*B[1,1]} = {E[0,1]}")
print(f"E[1,0] = {A[1,0]}*{B[0,0]} + {A[1,1]}*{B[1,0]} = {A[1,0]*B[0,0]} + {A[1,1]*B[1,0]} = {E[1,0]}")
print(f"E[1,1] = {A[1,0]}*{B[0,1]} + {A[1,1]}*{B[1,1]} = {A[1,0]*B[0,1]} + {A[1,1]*B[1,1]} = {E[1,1]}")
```
