## í•™ìŠµ ëª©í‘œ
- ì—¬ëŸ¬ DataFrameì„ í•˜ë‚˜ë¡œ í•©ì¹˜ëŠ” `concat`ê³¼ `merge` ì‚¬ìš©ë²• ìµíˆê¸°
- SQLì˜ Join ê°œë…ê³¼ Pandasì˜ `merge` ë¹„êµ
- ë°ì´í„° êµ¬ì¡°ë¥¼ ë°”ê¾¸ëŠ” `melt` ì´í•´í•˜ê¸°

````python
import pandas as pd

# ì˜ˆì œ ë°ì´í„° ìƒì„±
df1 = pd.DataFrame({'ID': [1, 2, 3], 'Name': ['Alice', 'Bob', 'Charlie']})
df2 = pd.DataFrame({'ID': [4, 5], 'Name': ['David', 'Eve']})

display(df1)
display(df2)
````

## ğŸ“š 1. ì´ì–´ ë¶™ì´ê¸° (`concat`)

ë¬¼ë¦¬ì ìœ¼ë¡œ ë°ì´í„°ë¥¼ ìœ„ì•„ë˜(í–‰) ë˜ëŠ” ì–‘ì˜†(ì—´)ìœ¼ë¡œ ë¶™ì…ë‹ˆë‹¤.

````python
# ìœ„ì•„ë˜ë¡œ í•©ì¹˜ê¸° (ê¸°ë³¸ê°’ axis=0)
df_concat = pd.concat([df1, df2], ignore_index=True)
print("ìœ„ì•„ë˜ë¡œ í•©ì¹˜ê¸°:")
display(df_concat)
````

## ğŸ“š 2. ë³‘í•©í•˜ê¸° (`merge`)

ê³µí†µëœ í‚¤(Key)ë¥¼ ê¸°ì¤€ìœ¼ë¡œ ë°ì´í„°ë¥¼ í•©ì¹©ë‹ˆë‹¤. SQLì˜ Joinê³¼ ê°™ìŠµë‹ˆë‹¤.

- **Inner Join**: êµì§‘í•© (ê³µí†µëœ í‚¤ë§Œ ë‚¨ê¹€)
- **Left Join**: ì™¼ìª½ ë°ì´í„° ëª¨ë‘ ìœ ì§€ + ì˜¤ë¥¸ìª½ ë§¤ì¹­ë˜ëŠ” ê²ƒ
- **Outer Join**: í•©ì§‘í•© (ëª¨ë“  ë°ì´í„° ìœ ì§€)

````python
# ì§ì› ì •ë³´
df_emp = pd.DataFrame({'ID': [1, 2, 3, 4], 'Name': ['Alice', 'Bob', 'Charlie', 'David']})
# ë¶€ì„œ ì •ë³´ (ID 3ë²ˆ CharlieëŠ” ë¶€ì„œ ì •ë³´ ì—†ìŒ, ID 5ë²ˆ EveëŠ” ì§ì› ì •ë³´ ì—†ìŒ)
df_dept = pd.DataFrame({'ID': [1, 2, 4, 5], 'Department': ['HR', 'Eng', 'Sales', 'HR']})

print("ì§ì›:")
display(df_emp)
print("ë¶€ì„œ:")
display(df_dept)
````

````python
# Inner Join (ê³µí†µ IDì¸ 1, 2, 4ë§Œ ë‚¨ìŒ)
merge_inner = pd.merge(df_emp, df_dept, on='ID', how='inner')
print("Inner Join:")
display(merge_inner)

# Left Join (ì§ì› ì •ë³´ ëª¨ë‘ ìœ ì§€, ë¶€ì„œ ì—†ìœ¼ë©´ NaN)
merge_left = pd.merge(df_emp, df_dept, on='ID', how='left')
print("\nLeft Join:")
display(merge_left)
````

## ğŸ“š 3. í˜•íƒœ ë³€ê²½ (`melt`)

Wide Format(ê°€ë¡œë¡œ ê¸´ ë°ì´í„°)ì„ Long Format(ì„¸ë¡œë¡œ ê¸´ ë°ì´í„°)ìœ¼ë¡œ ë°”ê¿€ ë•Œ ìœ ìš©í•©ë‹ˆë‹¤. ì‹œê°í™”í•  ë•Œ ìì£¼ ì“°ì…ë‹ˆë‹¤.

````python
df_wide = pd.DataFrame({
    'Name': ['Alice', 'Bob'],
    'Math': [90, 80],
    'English': [85, 95]
})
print("Wide Format:")
display(df_wide)

# Melt: ê³¼ëª©(Subject)ê³¼ ì ìˆ˜(Score)ë¡œ ë³€í™˜
df_long = pd.melt(df_wide, id_vars=['Name'], var_name='Subject', value_name='Score')
print("\nLong Format (Melted):")
display(df_long)
````
