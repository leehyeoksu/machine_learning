## í•™ìŠµ ëª©í‘œ
- DataFrameì—ì„œ ì›í•˜ëŠ” í–‰ê³¼ ì—´ì„ ì„ íƒí•˜ëŠ” ë°©ë²• (`loc`, `iloc`)
- ì¡°ê±´ì— ë§žëŠ” ë°ì´í„°ë§Œ ì¶”ì¶œí•˜ëŠ” ë¶ˆë¦¬ì–¸ ì¸ë±ì‹±(Boolean Indexing)
- íŠ¹ì • ê°’ì´ í¬í•¨ëœ ë°ì´í„°ë¥¼ ì°¾ëŠ” `isin` ë©”ì„œë“œ í™œìš©

```python
import pandas as pd

# ì‹¤ìŠµìš© ë°ì´í„° ìƒì„±
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
    'Age': [25, 30, 35, 40, 22],
    'City': ['New York', 'Los Angeles', 'Chicago', 'Houston', 'New York'],
    'Salary': [5000, 6000, 7000, 8000, 4500]
}
df = pd.DataFrame(data)
display(df)
```

## ðŸ“š 1. ì—´(Column) ì„ íƒ

ê°€ìž¥ ê¸°ë³¸ì ì¸ ì„ íƒ ë°©ë²•ìž…ë‹ˆë‹¤.

```python
# ë‹¨ì¼ ì—´ ì„ íƒ (Series ë°˜í™˜)
print("ì´ë¦„ ì—´:")
print(df['Name'])

# ì—¬ëŸ¬ ì—´ ì„ íƒ (DataFrame ë°˜í™˜ - ëŒ€ê´„í˜¸ 2ê°œ [[ ]])
print("\nì´ë¦„ê³¼ ì—°ë´‰ ì—´:")
display(df[['Name', 'Salary']])
```

## ðŸ“š 2. í–‰(Row) ì„ íƒ: `loc` vs `iloc`

Pandas ë°ì´í„° ì„ íƒì˜ í•µì‹¬ìž…ë‹ˆë‹¤.

| êµ¬ë¶„ | `loc` (Label-based) | `iloc` (Integer-based) |
|------|---------------------|------------------------|
| ê¸°ì¤€ | **ì´ë¦„(Label)** | **ìˆœì„œ(Index Number)** |
| ë²”ìœ„ | `start:end` (**end í¬í•¨**) | `start:end` (**end ë¯¸í¬í•¨**) |

### 2.1 `loc` ì‚¬ìš©ë²•
`df.loc[í–‰, ì—´]`

```python
# 0ë²ˆ í–‰(ì¸ë±ìŠ¤ ì´ë¦„ì´ 0ì¸ í–‰) ì„ íƒ
print("0ë²ˆ í–‰:")
print(df.loc[0])

# 0ë²ˆë¶€í„° 2ë²ˆ í–‰ê¹Œì§€, 'Name'ê³¼ 'City' ì—´ ì„ íƒ
# ì£¼ì˜: locì˜ ìŠ¬ë¼ì´ì‹±ì€ ëê°’ì„ í¬í•¨í•©ë‹ˆë‹¤!
print("\n0~2ë²ˆ í–‰, Name/City ì—´:")
display(df.loc[0:2, ['Name', 'City']])
```

### 2.2 `iloc` ì‚¬ìš©ë²•
`df.iloc[í–‰ë²ˆí˜¸, ì—´ë²ˆí˜¸]`

```python
# 0ë²ˆì§¸ í–‰ ì„ íƒ
print("0ë²ˆì§¸ í–‰:")
print(df.iloc[0])

# 0ë¶€í„° 2ë²ˆ ì „ê¹Œì§€(0, 1) í–‰, 0ë¶€í„° 2ë²ˆ ì „ê¹Œì§€(0, 1) ì—´
# íŒŒì´ì¬ ë¦¬ìŠ¤íŠ¸ ìŠ¬ë¼ì´ì‹±ê³¼ ë™ì¼í•˜ê²Œ ëê°’ì„ í¬í•¨í•˜ì§€ ì•ŠìŠµë‹ˆë‹¤.
print("\n0~1ë²ˆ í–‰, 0~1ë²ˆ ì—´:")
display(df.iloc[0:2, 0:2])
```

## ðŸ“š 3. ì¡°ê±´ë¶€ ì„ íƒ (Boolean Indexing)

íŠ¹ì • ì¡°ê±´ì„ ë§Œì¡±í•˜ëŠ” ë°ì´í„°ë§Œ í•„í„°ë§í•  ë•Œ ì‚¬ìš©í•©ë‹ˆë‹¤.

```python
# ë‚˜ì´ê°€ 30 ì´ìƒì¸ ì‚¬ëžŒ
mask = df['Age'] >= 30
print("ì¡°ê±´ ë§ˆìŠ¤í¬:")
print(mask)

print("\ní•„í„°ë§ ê²°ê³¼:")
display(df[mask]) # ë˜ëŠ” df[df['Age'] >= 30]
```

```python
# ë³µí•© ì¡°ê±´ (AND: &, OR: |)
# ë‚˜ì´ê°€ 30 ì´ìƒì´ê³ (AND), ë„ì‹œê°€ 'New York'ì´ ì•„ë‹Œ ì‚¬ëžŒ
condition = (df['Age'] >= 30) & (df['City'] != 'New York')
display(df[condition])
```

## ðŸ“š 4. ìœ ìš©í•œ í•„í„°ë§ í•¨ìˆ˜

- `isin()`: íŠ¹ì • ê°’ ë¦¬ìŠ¤íŠ¸ì— í¬í•¨ë˜ëŠ”ì§€ í™•ì¸
- `str.contains()`: ë¬¸ìžì—´ í¬í•¨ ì—¬ë¶€ í™•ì¸

```python
# isin ì‚¬ìš©
# ë„ì‹œê°€ New York ë˜ëŠ” Chicagoì¸ ì‚¬ëžŒ
cities = ['New York', 'Chicago']
display(df[df['City'].isin(cities)])
```

```python
# str.contains ì‚¬ìš©
# ì´ë¦„ì— 'a'ê°€ ë“¤ì–´ê°€ëŠ” ì‚¬ëžŒ (ëŒ€ì†Œë¬¸ìž êµ¬ë¶„)
display(df[df['Name'].str.contains('a')])
```

## ðŸ”¥ ì—°ìŠµ ë¬¸ì œ

1. `Salary`ê°€ 6000 ì´ìƒì¸ ì‚¬ëžŒë“¤ì˜ `Name`ê³¼ `City`ë§Œ ì„ íƒí•˜ì—¬ ì¶œë ¥í•˜ì„¸ìš”.
2. `City`ê°€ 'New York'ì´ë©´ì„œ `Age`ê°€ 25 ì´í•˜ì¸ ì‚¬ëžŒì„ ì°¾ì•„ ì¶œë ¥í•˜ì„¸ìš”.

```python
# ì •ë‹µ ì½”ë“œ

# 1. Salary >= 6000, Name/City ì„ íƒ
print("ë¬¸ì œ 1 ì •ë‹µ:")
result1 = df.loc[df['Salary'] >= 6000, ['Name', 'City']]
display(result1)

# 2. City == 'New York' AND Age <= 25
print("\në¬¸ì œ 2 ì •ë‹µ:")
result2 = df[(df['City'] == 'New York') & (df['Age'] <= 25)]
display(result2)
```
