````python
# ë¦¬ìŠ¤íŠ¸ë¡œ Series ìƒì„±
s = pd.Series([10, 20, 30, 40])
print("ê¸°ë³¸ Series:")
print(s)
print("\nValues:", s.values)
print("Index:", s.index)
````

### 2.2 DataFrame (ë°ì´í„°í”„ë ˆì„)

````python
# 1. ë”•ì…”ë„ˆë¦¬ë¡œ ìƒì„± (ê°€ì¥ í”í•œ ë°©ë²•)
data = {
    'Name': ['Alice', 'Bob', 'Charlie'],
    'Age': [25, 30, 35],
    'City': ['New York', 'Paris', 'London']
}
df = pd.DataFrame(data)
print("ë”•ì…”ë„ˆë¦¬ë¡œ ìƒì„±:")
display(df)
````
