## í•™ìŠµ ëª©í‘œ
- ê²°ì¸¡ì¹˜(Missing Data)ë¥¼ í™•ì¸í•˜ê³  ì²˜ë¦¬í•˜ëŠ” ë°©ë²• (`dropna`, `fillna`)
- ë°ì´í„° ì •ë ¬ (`sort_values`) ë° ì»¬ëŸ¼ ì¡°ì‘
- í•¨ìˆ˜ ì ìš© (`apply`, `map`) ë° ë¬¸ìì—´ ì²˜ë¦¬

```python
import pandas as pd
import numpy as np

# ê²°ì¸¡ì¹˜ê°€ í¬í•¨ëœ ë°ì´í„° ìƒì„±
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
    'Age': [25, np.nan, 35, 40, 22],
    'City': ['New York', 'Los Angeles', np.nan, 'Houston', 'New York'],
    'Salary': [5000, 6000, 7000, np.nan, 4500]
}
df = pd.DataFrame(data)
display(df)
```

## ğŸ“š 1. ê²°ì¸¡ì¹˜ ì²˜ë¦¬ (Missing Data)

ë°ì´í„° ë¶„ì„ì—ì„œ ê°€ì¥ í”í•˜ê³  ì¤‘ìš”í•œ ë¬¸ì œì…ë‹ˆë‹¤. `NaN` (Not a Number)ìœ¼ë¡œ í‘œì‹œë©ë‹ˆë‹¤.

```python
# ê²°ì¸¡ì¹˜ í™•ì¸
print("ê²°ì¸¡ì¹˜ ì—¬ë¶€ (True/False):")
display(df.isna())

print("\nì»¬ëŸ¼ë³„ ê²°ì¸¡ì¹˜ ê°œìˆ˜:")
print(df.isna().sum())
```

```python
# 1. ê²°ì¸¡ì¹˜ ì œê±° (dropna)
# í•˜ë‚˜ë¼ë„ NaNì´ ìˆìœ¼ë©´ í•´ë‹¹ í–‰ ì‚­ì œ
df_dropped = df.dropna()
print("ê²°ì¸¡ì¹˜ ì œê±° í›„:")
display(df_dropped)
```

```python
# 2. ê²°ì¸¡ì¹˜ ì±„ìš°ê¸° (fillna)
# AgeëŠ” í‰ê· ìœ¼ë¡œ, CityëŠ” 'Unknown'ìœ¼ë¡œ, SalaryëŠ” 0ìœ¼ë¡œ ì±„ìš°ê¸°
fill_values = {
    'Age': df['Age'].mean(),
    'City': 'Unknown',
    'Salary': 0
}
df_filled = df.fillna(fill_values)
print("ê²°ì¸¡ì¹˜ ì±„ìš´ í›„:")
display(df_filled)
```

## ğŸ“š 2. ë°ì´í„° ì •ë ¬ ë° ì¡°ì‘


```python
# ì •ë ¬ (sort_values)
# ë‚˜ì´ ê¸°ì¤€ ë‚´ë¦¼ì°¨ìˆœ ì •ë ¬
df_sorted = df_filled.sort_values(by='Age', ascending=False)
print("ë‚˜ì´ ë‚´ë¦¼ì°¨ìˆœ ì •ë ¬:")
display(df_sorted)
```

```python
# ì»¬ëŸ¼ ì¶”ê°€ ë° ì—°ì‚°
# ì—°ë´‰ ì¸ìƒ (10%)
df_filled['Salary_Up'] = df_filled['Salary'] * 1.1
display(df_filled)
```

## ğŸ“š 3. í•¨ìˆ˜ ì ìš© (`apply`)

ë°ì´í„°í”„ë ˆì„ì˜ í–‰ì´ë‚˜ ì—´ì— ë³µì¡í•œ í•¨ìˆ˜ë¥¼ ì ìš©í•  ë•Œ ì‚¬ìš©í•©ë‹ˆë‹¤.

```python
# ë‚˜ì´ëŒ€ë¥¼ êµ¬ë¶„í•˜ëŠ” í•¨ìˆ˜
def get_age_group(age):
    if age < 30:
        return 'Young'
    elif age < 40:
        return 'Middle'
    else:
        return 'Senior'

# 'Age' ì—´ì— í•¨ìˆ˜ ì ìš©
df_filled['Age_Group'] = df_filled['Age'].apply(get_age_group)
display(df_filled)
```

## ğŸ”¥ ì—°ìŠµ ë¬¸ì œ

1. `Salary`ì˜ ê²°ì¸¡ì¹˜ë¥¼ `Salary`ì˜ ì¤‘ì•™ê°’(median)ìœ¼ë¡œ ì±„ìš°ì„¸ìš”.
2. `City` ì—´ì˜ ëª¨ë“  ë¬¸ìì—´ì„ ëŒ€ë¬¸ìë¡œ ë°”ê¾¸ì„¸ìš”. (íŒíŠ¸: `df['City'].str.upper()` ì‚¬ìš©)

```python
# ì •ë‹µ ì½”ë“œ

# 1. ì¤‘ì•™ê°’ìœ¼ë¡œ ì±„ìš°ê¸°
df_median = df.copy()
df_median['Salary'] = df_median['Salary'].fillna(df_median['Salary'].median())
print("Salary ì¤‘ì•™ê°’ ì±„ì›€:")
display(df_median)

# 2. ëŒ€ë¬¸ì ë³€í™˜
# ì£¼ì˜: NaNì´ ìˆëŠ” ìƒíƒœì—ì„œ str ì ‘ê·¼ìë¥¼ ì“°ë©´ NaNì€ ê·¸ëŒ€ë¡œ ìœ ì§€ë¨
df_median['City'] = df_median['City'].str.upper()
print("\nCity ëŒ€ë¬¸ì ë³€í™˜:")
display(df_median)
```
