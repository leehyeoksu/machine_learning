## ğŸ“š 1. Reshape - ë°°ì—´ í˜•íƒœ ë³€ê²½

### 1.1 ê¸°ë³¸ reshape

```python
# 1ì°¨ì› ë°°ì—´ ìƒì„±
arr = np.arange(12)  # 0ë¶€í„° 11ê¹Œì§€
print("ì›ë³¸ ë°°ì—´:")
print(arr)
print(f"Shape: {arr.shape}")
print()
```

```python
# 3x4 í–‰ë ¬ë¡œ reshape
reshaped_3x4 = arr.reshape(3, 4)
print("3x4ë¡œ reshape:")
print(reshaped_3x4)
print(f"Shape: {reshaped_3x4.shape}")
print()

# ë°ì´í„°ëŠ” í–‰ ê¸°ì¤€(row-major)ìœ¼ë¡œ ì±„ì›Œì§‘ë‹ˆë‹¤
# [0, 1, 2, 3]     â† ì²« ë²ˆì§¸ í–‰
# [4, 5, 6, 7]     â† ë‘ ë²ˆì§¸ í–‰
# [8, 9, 10, 11]   â† ì„¸ ë²ˆì§¸ í–‰
```

```python
# 2x6 í–‰ë ¬ë¡œ reshape
reshaped_2x6 = arr.reshape(2, 6)
print("2x6ë¡œ reshape:")
print(reshaped_2x6)
print(f"Shape: {reshaped_2x6.shape}")
print()
```

```python
# 3ì°¨ì› ë°°ì—´ë¡œ reshape
reshaped_3d = arr.reshape(2, 2, 3)
print("2x2x3 3ì°¨ì› ë°°ì—´ë¡œ reshape:")
print(reshaped_3d)
print(f"Shape: {reshaped_3d.shape}")
print()

print("í•´ì„: 2ê°œì˜ (2x3) í–‰ë ¬")
print("ì²« ë²ˆì§¸ 2x3:")
print(reshaped_3d[0])
print("\në‘ ë²ˆì§¸ 2x3:")
print(reshaped_3d[1])
```

### 1.2 -1ì„ ì´ìš©í•œ ìë™ ê³„ì‚°

```python
# -1ì„ ì‚¬ìš©í•˜ë©´ ë‚¨ì€ ì°¨ì›ì„ ìë™ìœ¼ë¡œ ê³„ì‚°
arr = np.arange(24)

# 4í–‰ìœ¼ë¡œ í•˜ê³  ì—´ì€ ìë™ ê³„ì‚°
auto_reshape = arr.reshape(4, -1)
print("reshape(4, -1):")
print(auto_reshape)
print(f"Shape: {auto_reshape.shape}  # 24 Ã· 4 = 6 (ìë™ê³„ì‚°)")
print()
```

```python
# 3ì°¨ì›ì—ì„œë„ ì‚¬ìš© ê°€ëŠ¥
auto_3d = arr.reshape(2, 3, -1)
print("reshape(2, 3, -1):")
print(auto_3d)
print(f"Shape: {auto_3d.shape}  # 24 Ã· (2Ã—3) = 4 (ìë™ê³„ì‚°)")
print()

# -1ë¡œ 1ì°¨ì›ìœ¼ë¡œ í´ê¸° (flattenê³¼ ìœ ì‚¬)
flattened = arr.reshape(-1)
print("reshape(-1): 1ì°¨ì›ìœ¼ë¡œ í´ê¸°")
print(flattened)
print(f"Shape: {flattened.shape}")
```

### ğŸ’¬ Reshape ì£¼ì˜ì‚¬í•­
- reshape ê²°ê³¼ëŠ” ì›ë³¸ ë°°ì—´ê³¼ ë©”ëª¨ë¦¬ë¥¼ ê³µìœ í•  ìˆ˜ ìˆìŒ (view)
- ì›ë³¸ ë³€ê²½ ì‹œ reshapeëœ ë°°ì—´ë„ ì˜í–¥ì„ ë°›ì„ ìˆ˜ ìˆìŒ
- ë…ë¦½ì ì¸ ë³µì‚¬ê°€ í•„ìš”í•˜ë©´ `.copy()` ì‚¬ìš©

```python
# View vs Copy ì˜ˆì œ
original = np.arange(6)
view = original.reshape(2, 3)  # view (ë©”ëª¨ë¦¬ ê³µìœ )
copy = original.reshape(2, 3).copy()  # ë…ë¦½ì  ë³µì‚¬

print("ì›ë³¸:", original)
print("View:", view)
print("Copy:", copy)
print()

# ì›ë³¸ ìˆ˜ì •
original[0] = 999

print("ì›ë³¸ ìˆ˜ì • í›„:")
print("ì›ë³¸:", original)
print("View:", view, " â† ë³€ê²½ë¨!")
print("Copy:", copy, " â† ë³€ê²½ ì•ˆë¨")
```

## ğŸ“š 2. Transpose - ì „ì¹˜

### 2.1 2ì°¨ì› ë°°ì—´ì˜ ì „ì¹˜

```python
# 2x3 í–‰ë ¬ ìƒì„±
M = np.array([[1, 2, 3],
              [4, 5, 6]])

print("ì›ë³¸ í–‰ë ¬ (2x3):")
print(M)
print(f"Shape: {M.shape}")
print()
```

```python
# ì „ì¹˜ (í–‰ê³¼ ì—´ ë°”ê¾¸ê¸°)
M_T = M.T  # ë˜ëŠ” np.transpose(M)

print("ì „ì¹˜ í–‰ë ¬ (3x2):")
print(M_T)
print(f"Shape: {M_T.shape}")
print()

print("ì „ì¹˜ ê³¼ì •:")
print("M[0,0]=1 â†’ M_T[0,0]=1")
print("M[0,1]=2 â†’ M_T[1,0]=2")
print("M[0,2]=3 â†’ M_T[2,0]=3")
print("M[1,0]=4 â†’ M_T[0,1]=4")
print("M[1,1]=5 â†’ M_T[1,1]=5")
print("M[1,2]=6 â†’ M_T[2,1]=6")
```

### 2.2 ë‹¤ì°¨ì› ë°°ì—´ì˜ ì¶• êµí™˜

```python
# 3ì°¨ì› ë°°ì—´ ìƒì„± (ë°°ì¹˜, ë†’ì´, ë„ˆë¹„)
arr_3d = np.arange(24).reshape(2, 3, 4)
print("ì›ë³¸ 3D ë°°ì—´ (2, 3, 4):")
print(arr_3d)
print(f"Shape: {arr_3d.shape}")
print()
```

```python
# ì¶• ìˆœì„œ ë³€ê²½: (0, 1, 2) â†’ (2, 0, 1)
transposed = arr_3d.transpose(2, 0, 1)
print("transpose(2, 0, 1) â†’ Shape (4, 2, 3):")
print(transposed)
print(f"Shape: {transposed.shape}")
print()

# ì¶• ì„¤ëª…:
# ì›ë³¸: axis 0=2, axis 1=3, axis 2=4
# ê²°ê³¼: axis 0=4(ì›ë˜ 2), axis 1=2(ì›ë˜ 0), axis 2=3(ì›ë˜ 1)
```

```python
# ì´ë¯¸ì§€ ë°ì´í„° ì°¨ì› ë³€ê²½ ì˜ˆì œ
# (ë†’ì´, ë„ˆë¹„, ì±„ë„) â†’ (ì±„ë„, ë†’ì´, ë„ˆë¹„)
image_hwc = np.random.rand(28, 28, 3)  # 28x28 RGB ì´ë¯¸ì§€
print(f"HWC í˜•ì‹: {image_hwc.shape}  # (ë†’ì´, ë„ˆë¹„, ì±„ë„)")

image_chw = image_hwc.transpose(2, 0, 1)
print(f"CHW í˜•ì‹: {image_chw.shape}  # (ì±„ë„, ë†’ì´, ë„ˆë¹„)")
print("â€» PyTorchëŠ” CHW í˜•ì‹ì„ ì„ í˜¸í•©ë‹ˆë‹¤")
```

## ğŸ“š 4. ë°°ì—´ ì—°ê²° (Concatenate, Stack)

### 4.1 Concatenate - ê¸°ì¡´ ì°¨ì›ìœ¼ë¡œ ì—°ê²°

```python
# 1ì°¨ì› ë°°ì—´ ì—°ê²°
import numpy as np
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])

concat = np.concatenate((a, b) )
print(f"{a.shape}\n")
print(b.shape)
print("1D ë°°ì—´ concatenate:")
print(f"a = {a}")
print(f"b = {b}")
print(f"concatenate([a, b]) = {concat}")
print(f"Shape: {concat.shape}")
print()
```

```python
# 2ì°¨ì› ë°°ì—´ ì—°ê²° - axis=0 (í–‰ ë°©í–¥)
A = np.array([[1, 2],
              [3, 4]])
B = np.array([[5, 6],
              [7, 8]])

concat_axis0 = np.concatenate([A, B], axis=0)
print("2D ë°°ì—´ concatenate (axis=0, í–‰ ë°©í–¥):")
print(concat_axis0)
print(f"Shape: {concat_axis0.shape}  # (4, 2)")
print()
```
