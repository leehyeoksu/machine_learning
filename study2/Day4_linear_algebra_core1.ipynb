## ğŸ“š 1. ì—­í–‰ë ¬ ê³„ì‚°

### 1.1 2Ã—2 í–‰ë ¬ì˜ ì—­í–‰ë ¬

```python
# 2Ã—2 í–‰ë ¬ì˜ ì—­í–‰ë ¬
import numpy as np
A = np.array([[4, 7],
              [2, 6]])

print('í–‰ë ¬ A:')
print(A)
print()

# NumPyë¡œ ì—­í–‰ë ¬ ê³„ì‚°
A_inv = np.linalg.inv(A)
print('ì—­í–‰ë ¬ Aâ»Â¹:')
print(A_inv)
print()

# ìˆ˜ë™ ê³„ì‚° (2Ã—2 ê³µì‹)
# Aâ»Â¹ = 1/det(A) * [[d, -b], [-c, a]]
det_A = 4*6 - 7*2
print(f'í–‰ë ¬ì‹: det(A) = 4Ã—6 - 7Ã—2 = {det_A}')
print()

A_inv_manual = (1/det_A) * np.array([[6, -7], [-2, 4]])
print('ìˆ˜ë™ ê³„ì‚°í•œ ì—­í–‰ë ¬:')
print(A_inv_manual)
print()

print(f'NumPy ê²°ê³¼ì™€ ì¼ì¹˜: {np.allclose(A_inv, A_inv_manual)}')
```

```python
# ì—­í–‰ë ¬ ê²€ì¦: A @ Aâ»Â¹ = I
import numpy as np
identity = A @ A_inv
print('A @ Aâ»Â¹:')
print(identity)
print()

# ë‹¨ìœ„í–‰ë ¬ê³¼ ë¹„êµ
I = np.eye(2)
print('ë‹¨ìœ„í–‰ë ¬ I:')
print(I)
print()

print(f'ë‹¨ìœ„í–‰ë ¬ê³¼ ì¼ì¹˜ (í—ˆìš©ì˜¤ì°¨ ê³ ë ¤): {np.allclose(identity, I)}')
```

### 1.2 3Ã—3 í–‰ë ¬ì˜ ì—­í–‰ë ¬

```python
# 3Ã—3 í–‰ë ¬
B = np.array([[1, 2, 3],
              [0, 1, 4],
              [5, 6, 0]])

print('í–‰ë ¬ B:')
print(B)
print()

# ì—­í–‰ë ¬ ê³„ì‚°
B_inv = np.linalg.inv(B)
print('ì—­í–‰ë ¬ Bâ»Â¹:')
print(B_inv)
print()

# ê²€ì¦
identity_3 = B @ B_inv
print('B @ Bâ»Â¹:')
print(identity_3)
print()

# ì‹¤ìˆ˜ ì˜¤ì°¨ í™•ì¸
print(f'ë‹¨ìœ„í–‰ë ¬ê³¼ ì¼ì¹˜ (í—ˆìš©ì˜¤ì°¨): {np.allclose(identity_3, np.eye(3))}')
print(f'ìµœëŒ€ ì˜¤ì°¨: {np.max(np.abs(identity_3 - np.eye(3))):.2e}')
```

## ğŸ“š 2. í–‰ë ¬ì‹ (Determinant)

### 2.1 í–‰ë ¬ì‹ ê³„ì‚°

```python
# 2Ã—2 í–‰ë ¬ì˜ í–‰ë ¬ì‹
A = np.array([[3, 8],
              [4, 6]])

det_A = np.linalg.det(A)
print('í–‰ë ¬ A:')
print(A)
print()

print(f'det(A) = {det_A}')
print()

# ìˆ˜ë™ ê³„ì‚°
det_manual = 3*6 - 8*4
print(f'ìˆ˜ë™ ê³„ì‚°: 3Ã—6 - 8Ã—4 = {det_manual}')
print(f'ì¼ì¹˜: {np.isclose(det_A, det_manual)}')
```

```python
# 3Ã—3 í–‰ë ¬ì˜ í–‰ë ¬ì‹
C = np.array([[6, 1, 1],
              [4, -2, 5],
              [2, 8, 7]])

det_C = np.linalg.det(C)
print('í–‰ë ¬ C:')
print(C)
print(f'\ndet(C) = {det_C:.2f}')
```

### 2.2 íŠ¹ì´í–‰ë ¬ (Singular Matrix)

```python
# í–‰ë ¬ì‹ì´ 0ì¸ íŠ¹ì´í–‰ë ¬ (ì—­í–‰ë ¬ ì—†ìŒ)
S = np.array([[1, 2, 3],
              [2, 4, 6],
              [3, 6, 9]])

print('íŠ¹ì´í–‰ë ¬ S (ë‘ ë²ˆì§¸ í–‰ = ì²« ë²ˆì§¸ í–‰ Ã— 2):')
print(S)
print()

det_S = np.linalg.det(S)
print(f'det(S) = {det_S:.2e}  (0ì— ë§¤ìš° ê°€ê¹Œì›€)')
print()

# ì—­í–‰ë ¬ ì‹œë„
try:
    S_inv = np.linalg.inv(S)
    print('ì—­í–‰ë ¬ (ë¶ˆì•ˆì •):', S_inv)
except np.linalg.LinAlgError as e:
    print(f'âŒ ì˜¤ë¥˜: {e}')
    print('ì„¤ëª…: det(S)=0ì´ë¯€ë¡œ ì—­í–‰ë ¬ì´ ì¡´ì¬í•˜ì§€ ì•ŠìŠµë‹ˆë‹¤')
```

## ğŸ“š 4. ì„ í˜• ë°©ì •ì‹ í’€ì´

### 4.1 np.linalg.solve ì‚¬ìš©

```python
# ì—°ë¦½ë°©ì •ì‹ Ax = b í’€ì´
# 2x + 3y = 8
# x - 4y = -2

A = np.array([[2, 3],
              [1, -4]])
b = np.array([8, -2])

print('ì—°ë¦½ë°©ì •ì‹:')
print('2x + 3y = 8')
print('x - 4y = -2')
print()

# ë°©ë²• 1: np.linalg.solve
x = np.linalg.solve(A, b)
print('í•´ (np.linalg.solve):')
print(f'x = {x[0]:.2f}, y = {x[1]:.2f}')
print()

# ê²€ì¦
result = A @ x
print('ê²€ì¦ (Ax):')
print(result)
print(f'bì™€ ì¼ì¹˜: {np.allclose(result, b)}')
print()
```

```python
# ë°©ë²• 2: ì—­í–‰ë ¬ ì‚¬ìš©
A_inv = np.linalg.inv(A)
x2 = A_inv @ b

print('í•´ (ì—­í–‰ë ¬ ì‚¬ìš©):')
print(f'x = {x2[0]:.2f}, y = {x2[1]:.2f}')
print()

print(f'ë‘ ë°©ë²•ì˜ ê²°ê³¼ ì¼ì¹˜: {np.allclose(x, x2)}')
print()
print('â€» ì‹¤ë¬´ì—ì„œëŠ” np.linalg.solveê°€ ë” íš¨ìœ¨ì ì´ê³  ì•ˆì •ì ì…ë‹ˆë‹¤')
```
