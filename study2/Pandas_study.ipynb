## ğŸ“š 1. ë°ì´í„° êµ¬ì¡° (Data Structures)

### 1.1 Series (ì‹œë¦¬ì¦ˆ)
- 1ì°¨ì› ë°ì´í„° êµ¬ì¡°
- ì¸ë±ìŠ¤(Index)ì™€ ê°’(Value)ìœ¼ë¡œ êµ¬ì„±

```python
# ë¦¬ìŠ¤íŠ¸ë¡œ Series ìƒì„±
s = pd.Series([10, 20, 30, 40, 50])
print("ê¸°ë³¸ Series:")
print(s)
print("\nê°’(Values):", s.values)
print("ì¸ë±ìŠ¤(Index):", s.index)
```

```python
# ì¸ë±ìŠ¤ë¥¼ ì§€ì •í•˜ì—¬ Series ìƒì„±
s_index = pd.Series([90, 80, 70], index=['Math', 'English', 'Science'])
print("ì¸ë±ìŠ¤ ì§€ì • Series:")
print(s_index)
print("\nMath ì ìˆ˜:", s_index['Math'])
```

### 1.2 DataFrame (ë°ì´í„°í”„ë ˆì„)
- 2ì°¨ì› ë°ì´í„° êµ¬ì¡° (í–‰ê³¼ ì—´)
- ì—¬ëŸ¬ ê°œì˜ Seriesê°€ ëª¨ì—¬ ìˆëŠ” í˜•íƒœ

```python
# ë”•ì…”ë„ˆë¦¬ë¡œ DataFrame ìƒì„±
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David'],
    'Age': [25, 30, 35, 40],
    'City': ['New York', 'Los Angeles', 'Chicago', 'Houston']
}

df = pd.DataFrame(data)
print("DataFrame ìƒì„±:")
display(df) # Jupyterì—ì„œëŠ” print(df) ëŒ€ì‹  display(df)ë¥¼ ì“°ë©´ ë” ì˜ˆì˜ê²Œ ë‚˜ì˜µë‹ˆë‹¤
```

ë°ì´í„°ë¥¼ ë¶ˆëŸ¬ì˜¨ í›„ ê°€ì¥ ë¨¼ì € í•´ì•¼ í•  ì¼ì€ ë°ì´í„°ì˜ ìƒê¹€ìƒˆì™€ êµ¬ì¡°ë¥¼ íŒŒì•…í•˜ëŠ” ê²ƒì…ë‹ˆë‹¤.

```python
# ìƒìœ„ 3ê°œ í–‰ í™•ì¸
print("Head (ìƒìœ„ 3ê°œ):")
display(df.head(3))

# í•˜ìœ„ 2ê°œ í–‰ í™•ì¸
print("\nTail (í•˜ìœ„ 2ê°œ):")
display(df.tail(2))

# ë°ì´í„°í”„ë ˆì„ ì •ë³´ í™•ì¸ (í–‰/ì—´ ê°œìˆ˜, ë°ì´í„° íƒ€ì…, ê²°ì¸¡ì¹˜ ë“±)
print("\nInfo:")
df.info()
```

## ğŸ“š 3. ë°ì´í„° ì„ íƒ (Selection)

ì›í•˜ëŠ” ë°ì´í„°ë¥¼ ì½• ì§‘ì–´ë‚´ëŠ” ë°©ë²•ì…ë‹ˆë‹¤. Pandasì—ì„œëŠ” `loc`ì™€ `iloc`ê°€ ê°€ì¥ ì¤‘ìš”í•©ë‹ˆë‹¤.

- **`loc` (Label-based)**: í–‰/ì—´ì˜ **ì´ë¦„(Label)**ìœ¼ë¡œ ì„ íƒ
- **`iloc` (Integer-based)**: í–‰/ì—´ì˜ **ì •ìˆ˜ ì¸ë±ìŠ¤(ìœ„ì¹˜)**ë¡œ ì„ íƒ

```python
# 1. ì—´(Column) ì„ íƒ
print("ì´ë¦„ ì—´ë§Œ ì„ íƒ:")
print(df['Name'])

# ì—¬ëŸ¬ ì—´ ì„ íƒ (ë¦¬ìŠ¤íŠ¸ë¡œ ì „ë‹¬)
print("\nì´ë¦„ê³¼ ë„ì‹œ ì„ íƒ:")
display(df[['Name', 'City']])
```

```python
# 2. loc ì‚¬ìš© (ì´ë¦„ìœ¼ë¡œ ì„ íƒ)
# í˜•ì‹: df.loc[í–‰, ì—´]

print("loc[0, 'Name']: ", df.loc[0, 'Name'])  # 0ë²ˆ ì¸ë±ìŠ¤ í–‰ì˜ 'Name' ì—´
print("\nloc[0:2, ['Name', 'Age']]:") # 0ë¶€í„° 2ê¹Œì§€(í¬í•¨!) í–‰
display(df.loc[0:2, ['Name', 'Age']])
```

```python
# 3. iloc ì‚¬ìš© (ìœ„ì¹˜ ì •ìˆ˜ë¡œ ì„ íƒ)
# í˜•ì‹: df.iloc[í–‰ë²ˆí˜¸, ì—´ë²ˆí˜¸]

print("iloc[0, 0]: ", df.iloc[0, 0])  # 0ë²ˆì§¸ í–‰, 0ë²ˆì§¸ ì—´
print("\niloc[0:2, 0:2]:") # 0ë¶€í„° 2ì „ê¹Œì§€(ë¯¸í¬í•¨!) í–‰
display(df.iloc[0:2, 0:2])
```

```python
# 4. ì¡°ê±´ë¶€ ì„ íƒ (Boolean Indexing)
# ë‚˜ì´ê°€ 30 ì´ìƒì¸ ì‚¬ëŒë§Œ ì„ íƒ
over_30 = df[df['Age'] >= 30]
print("ë‚˜ì´ê°€ 30 ì´ìƒì¸ í–‰:")
display(over_30)
```

## ğŸ“š 4. ë°ì´í„° ì¡°ì‘ (Manipulation)

ë°ì´í„°ë¥¼ ìˆ˜ì •í•˜ê³ , ì¶”ê°€í•˜ê³ , ì‚­ì œí•˜ëŠ” ë°©ë²•ì…ë‹ˆë‹¤.

```python
# ì—´ ì¶”ê°€
df['Country'] = 'USA'
print("Country ì—´ ì¶”ê°€ í›„:")
display(df)

# ì—´ ì‚­ì œ (drop)
# axis=0: í–‰ ì‚­ì œ, axis=1: ì—´ ì‚­ì œ
df_dropped = df.drop('Country', axis=1)
print("\nCountry ì—´ ì‚­ì œ í›„ (ì›ë³¸ ìœ ì§€):")
display(df_dropped)

# inplace=Trueë¥¼ ì“°ë©´ ì›ë³¸ì´ ë³€ê²½ë¨
# df.drop('Country', axis=1, inplace=True)
```

```python
# ì •ë ¬ (sort_values)
df_sorted = df.sort_values(by='Age', ascending=False) # ë‚˜ì´ ë‚´ë¦¼ì°¨ìˆœ
print("ë‚˜ì´ ë‚´ë¦¼ì°¨ìˆœ ì •ë ¬:")
display(df_sorted)
```

## ğŸ“š 5. ê²°ì¸¡ì¹˜ ì²˜ë¦¬ (Missing Data)

ì‹¤ì œ ë°ì´í„°ì—ëŠ” ê°’ì´ ë¹„ì–´ìˆëŠ” ê²½ìš°(NaN)ê°€ ë§ìŠµë‹ˆë‹¤. ì´ë¥¼ í™•ì¸í•˜ê³  ì²˜ë¦¬í•˜ëŠ” ê²ƒì€ ë§¤ìš° ì¤‘ìš”í•©ë‹ˆë‹¤.

```python
# ê²°ì¸¡ì¹˜ê°€ í¬í•¨ëœ ë°ì´í„° ìƒì„±
df_missing = df.copy()
df_missing.loc[1, 'Age'] = np.nan
df_missing.loc[3, 'City'] = np.nan

print("ê²°ì¸¡ì¹˜ê°€ ìˆëŠ” ë°ì´í„°:")
display(df_missing)

# ê²°ì¸¡ì¹˜ í™•ì¸ (isna)
print("\nê²°ì¸¡ì¹˜ ì—¬ë¶€ í™•ì¸:")
display(df_missing.isna())

# ê²°ì¸¡ì¹˜ ê°œìˆ˜ í™•ì¸
print("\nì—´ë³„ ê²°ì¸¡ì¹˜ ê°œìˆ˜:")
print(df_missing.isna().sum())
```

```python
# ê²°ì¸¡ì¹˜ ì±„ìš°ê¸° (fillna)
df_filled = df_missing.fillna({'Age': df_missing['Age'].mean(), 'City': 'Unknown'})
print("ê²°ì¸¡ì¹˜ ì±„ìš´ í›„ (ë‚˜ì´ëŠ” í‰ê· , ë„ì‹œëŠ” Unknown):")
display(df_filled)

# ê²°ì¸¡ì¹˜ ì œê±°í•˜ê¸° (dropna)
df_dropped_na = df_missing.dropna()
print("\nê²°ì¸¡ì¹˜ê°€ ìˆëŠ” í–‰ ì œê±° í›„:")
display(df_dropped_na)
```

## ğŸ“š 6. ê·¸ë£¹í™” ë° ì§‘ê³„ (Grouping & Aggregation)

ë°ì´í„°ë¥¼ íŠ¹ì • ê¸°ì¤€ìœ¼ë¡œ ë¬¶ì–´ì„œ í†µê³„ëŸ‰ì„ ê³„ì‚°í•  ë•Œ ì‚¬ìš©í•©ë‹ˆë‹¤. (`groupby`)

```python
# ì˜ˆì œ ë°ì´í„° í™•ì¥
df['Department'] = ['HR', 'Engineering', 'Engineering', 'HR']
df['Salary'] = [5000, 7000, 8000, 5500]
print("í™•ì¥ëœ ë°ì´í„°:")
display(df)

# ë¶€ì„œë³„ í‰ê·  ì—°ë´‰ ê³„ì‚°
dept_salary = df.groupby('Department')['Salary'].mean()
print("\në¶€ì„œë³„ í‰ê·  ì—°ë´‰:")
print(dept_salary)

# ì—¬ëŸ¬ í†µê³„ëŸ‰ í•œë²ˆì— ê³„ì‚° (agg)
dept_agg = df.groupby('Department')['Salary'].agg(['mean', 'sum', 'count'])
print("\në¶€ì„œë³„ ì—°ë´‰ í†µê³„:")
display(dept_agg)
```

## ğŸ“š 7. ë°ì´í„° ë³‘í•© (Merging)

ì—¬ëŸ¬ ê°œì˜ DataFrameì„ í•˜ë‚˜ë¡œ í•©ì¹˜ëŠ” ë°©ë²•ì…ë‹ˆë‹¤. (`concat`, `merge`)

```python
# ë°ì´í„°í”„ë ˆì„ 2ê°œ ìƒì„±
df1 = pd.DataFrame({'ID': [1, 2], 'Name': ['Alice', 'Bob']})
df2 = pd.DataFrame({'ID': [3, 4], 'Name': ['Charlie', 'David']})

# ìœ„ì•„ë˜ë¡œ ì´ì–´ ë¶™ì´ê¸° (concat)
df_concat = pd.concat([df1, df2], ignore_index=True)
print("concat ê²°ê³¼:")
display(df_concat)
```

```python
# í‚¤(Key)ë¥¼ ê¸°ì¤€ìœ¼ë¡œ ë³‘í•©í•˜ê¸° (merge)
df_info = pd.DataFrame({'ID': [1, 2, 3], 'Role': ['Manager', 'Developer', 'Designer']})

# df_concatê³¼ df_infoë¥¼ 'ID' ê¸°ì¤€ìœ¼ë¡œ ë³‘í•©
df_merged = pd.merge(df_concat, df_info, on='ID', how='left') # left join
print("merge ê²°ê³¼ (Left Join):")
display(df_merged)
```
