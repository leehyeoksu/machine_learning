```python
import numpy as np
import torch

rand=torch.rand([2,3])
print(rand)
```

```python
torch.manual_seed(42)
rand_int_x=torch.randint(low=1,high=10,size=(3,3))
rand_int_y=torch.randint(low=2,high=9,size=(3,3))
print(f"x의값:\n{rand_int_x}\n")
print(f"x의값:\n{rand_int_y}\n")
print(rand_int_x@rand_int_y.T)
```

```python
np.random.seed(42)
num=np.random.randint(1,10,(3,3))
num_float=np.random.rand(3,3)
print(f"{num}\n")
print(f"{num_float}\n")
```

```python
torch_refer_numpy=torch.from_numpy(num)
print(f"{torch_refer_numpy}\n")
print(f"{torch_refer_numpy.shape}")
```
