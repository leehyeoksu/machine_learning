```python
import numpy as np
import torch
np.random.seed(42)
a=np.random.randint(1,10,size=(4,4))
print(a)
flat=a.argmax()
row=a.argmax(axis=0)
col=a.argmax(axis=1)
print(f"행으로볼떄{row}")
print(f"열으로볼때{col}")
print(f"평탄화했을때{flat}")
print(a.sum())
print(a.sum(axis=0))
print(a.mean())
print(a.mean(axis=1))
print(a.std())
print(a.std(axis=0))
print(a.var())
print(a.var(axis=1))
```
