```python
import numpy as np 

num1=np.arange(0,9).reshape(3,3)
print(num1)
print(num1.shape)
print(num1.dtype)
print(num1.ndim)
print(num1.size)
u,s,vt=np.linalg.svd(num1)
ze=np.diag(s)
print(np.allclose(u@ze@vt, num1))
```
