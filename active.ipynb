```python
import torch
import numpy as np
np.random.seed(42)
a=np.random.randint(1,10,size=(4,4))
b=torch.from_numpy(a)
print(a)
a[:2,2:]=3
print(a)
a[2:,:2]=3
print(b)
print(torch.linalg.det(b.float()))
```
